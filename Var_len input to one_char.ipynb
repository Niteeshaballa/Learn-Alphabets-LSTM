```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.utils import np_utils
from keras.preprocessing.sequence import pad_sequences
```

```python
#seed
numpy.random.seed(7)
```

```python
#define raw dataset
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

#create mapping of characters to integers (0-25) and reverse
char_to_int = dict((c,i) for i,c in enumerate(alphabet))
int_to_char = dict((i,c) for i,c in enumerate(alphabet))
```

```python
#prepare dataset of input to output pairs encoded as integers
num_inputs=1000
max_len=5
dataX=[]
dataY=[]
for i in range(num_inputs):
    start = numpy.random.randint(len(alphabet)-2)
    end = numpy.random.randint(start, min(start+max_len, len(alphabet)-1))
    seq_in = alphabet[start: end+1]
    seq_out = alphabet[end+1]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
    print(seq_in ,'->', seq_out)
```

```python
#convert list of lists to array and pad sequences if needed
X = pad_sequences(dataX, maxlen=max_len, dtype='float32')
```

```python
#reshape X to be [samples,time steps, features]
X = numpy.reshape(X, (X.shape[0], max_len, 1))
```

```python
#normalize X as LSTM sigmoid needs between 0 & 1
X = X/float(len(alphabet))
```

```python
#one hot encoding of output
y = np_utils.to_categorical(dataY)
```

```python
#create and fit model
#sample naive LSTM model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], 1)))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=500, batch_size=1, verbose=1)
```

```python
#summarize performance of the model
scores = model.evaluate(X, y)
print("Accuracy: %.2f%%" %(scores[1]*100))
scores
```

```python
#demonstrate the model predicting random patterns
print("test a random pattern:")
for i in range(0,20):
    pattern_index = numpy.random.randint(len(dataX))
    pattern = dataX[pattern_index]
    x = pad_sequences([pattern], maxlen=max_len, dtype='float32')
    x = numpy.reshape(x, (1, max_len, 1))
    x = x/float(len(alphabet))
    prediction = model.predict(x)
    index = numpy.argmax(prediction)
    result = int_to_char[index]
    seq_in = [int_to_char[value] for value in pattern]
    print(seq_in ,'->', result)
```
